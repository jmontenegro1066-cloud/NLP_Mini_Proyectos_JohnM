```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report

# 1. Cargar el nuevo archivo de libros
df_libros = pd.read_csv('reseñas_libros.txt', sep='\t', header=None, names=['reseña', 'etiqueta'])

# 2. Análisis con VADER
nltk.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()

df_libros['puntajes'] = df_libros['reseña'].apply(lambda x: sid.polarity_scores(x))
df_libros['compuesto'] = df_libros['puntajes'].apply(lambda x: x['compound'])
df_libros['prediccion'] = df_libros['compuesto'].apply(lambda c: 1 if c > 0 else 0)

# 3. Visualización: Matriz de Confusión (VALOR AGREGADO)
cm = confusion_matrix(df_libros['etiqueta'], df_libros['prediccion'])
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicción de la IA')
plt.ylabel('Etiqueta Real')
plt.title('Matriz de Confusión: Reseñas de Libros')
plt.show()

print(classification_report(df_libros['etiqueta'], df_libros['prediccion']))
```
