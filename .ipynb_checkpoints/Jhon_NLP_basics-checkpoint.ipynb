```python
import spacy
import pandas as pd
import matplotlib.pyplot as plt

# 1. Cargar el modelo en español
nlp = spacy.load('es_core_news_sm')

# 2. LLAMADA AL ARCHIVO EXTERNO
with open('resumen_fisioterapia.txt', 'r', encoding='utf-8') as f:
    texto_salud = f.read()

doc = nlp(texto_salud)

# 3. Análisis Automático
data = []
for token in doc:
    if not token.is_punct and not token.is_space:
        data.append([token.text, token.pos_, token.lemma_])

df_analisis = pd.DataFrame(data, columns=['Palabra', 'Categoría', 'Lema'])

# 4. Hallazgo Personal y Visualización
print(f"Análisis de las primeras 5 palabras:\n{df_analisis.head()}")

df_analisis['Categoría'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Composición Gramatical del Texto de Fisioterapia")
plt.show()
```
